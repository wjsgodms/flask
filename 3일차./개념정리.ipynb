1. ORM의 의미와 사용 이유

orm이란 Object-Relationai Mapping을 말한다. 
객체 지향 프로그래밍 언어에서 객체를 사용해 관계형 데이터베이스와 상호장용하기 위한 기법이다.

    orm 사용 이유

    - 생산성 : SQL문을 직접 작성하지 않아도 돼서 코드 작성 시간이 줄어든다.
    - 가독성 : SQL문 대신 객체지향 코드를 작성하여 더 직관적이고 이해하기 쉽다. 
    - 유지보수 : 객체 지향 코드로 통합이 되어있어, 데이터 베이스 구조가 바뀌어도 쉽게 대응 가능하다.
    - 독립성 : ORM은 다양한 데이터베이스를 지원하여 코드 수정 없이 데이터베이스를 변경할 수 있다.
    - 보안 강화 : SQL injection 같은 보안 문제를 위한 메커니즘을 제공한다.

    

2. Model 정의하기

model은 데이터베이스 테이블의 구조를 코드로 표현한 것이다.

3. 모델 간의 관계 해석

ORM의 model간 관계는 데이터베이스의 테이블 간 관계와 같다.
1:1 관계, 1:N 과계 , N:M 관게가 있다.

    1:1 관계 - 한테이블의 레코드가 다른 테니블의 레코드와 하나씩만 연결되는 관계.
    1:N 관계 - 한테이블의 레코드가 다른 테이블의 여러 레코드와 연결되는 관계.
    N:M 관계 - 한테이블의 레코드가 다른 테이블의 여러 레코드와 연결되며, 반대로도 연결될 수 있는 관계.

4. 스키마와 모델 사이의 차이점

스키마는 데이터베이스의 원형설계도이고 모델은 이 설계도를 코드로 표현하고 제어하기 위한 도구이다.

6. 직렬화와 역직렬화의 의미

직결화와 역직결화는 데이터베이스 또는 ORM에서 모델 간 관계를 설정하거나 이해할 때 사용하는 개념이다.

    직결화

    - 직접 연결된 관계를 의미
    - 객체 또는 테이블 a가 b를 참조하는 관계


    역직결화

    - 관게가 역방향으로 설정됨
    - 객체 또는 테이블 b가 a를 참조하는 관계
